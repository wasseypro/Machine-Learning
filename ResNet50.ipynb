```
from keras.applications.resnet50 import ResNet50
```

```
# Import image module
from keras.preprocessing import image
```

```
# Import functions,preprocess_input,decode predictions
from keras.applications.resnet50 import preprocess_input,decode_predictions
```

```
# Capture weights from imagenet
model=ResNet50(weights='imagenet')
```

```
impg_path='/content/Res Images1/cat.4006.jpg'
```

```
img=image.load_img(impg_path,target_size=(224,224))
```

```
#Convert image into array
x=image.img_to_array(img)
```

```
import numpy as np
```

```
# expand the dimension
x=np.expand_dims(x,axis=0)
x=preprocess_input(x)
```

```
# to display the image
import matplotlib.pyplot as plt
```

```
img_path_img=plt.imread(impg_path)
plt.imshow(img_path_img)
plt.show()
```

```
preds=model.predict(x)
print(preds)
```

```
#decode the prediction
decode_predictions(preds,top=3)
```
